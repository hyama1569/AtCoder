```python
a, b = map(str, input().split())

pala = a[:2] + a[2] + a[:2][::-1]
palb = b[:2] + b[2] + b[:2][::-1]

heada = int(pala[:3])
headb = int(palb[:3])
if int(pala) < int(a):
    heada += 1

if int(palb) > int(b):
    headb -= 1

print(headb - heada + 1)
```
