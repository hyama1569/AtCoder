```python
def check_2(ls):
    flag = False
    for i in range(len(ls)):
        if ls[i] % 2 != 0:
            flag = True
    return flag
```

```python
n = int(input())
a_ls = list(map(int, input().split()))

def check_2(ls):
    flag = False
    for i in range(len(ls)):
        if ls[i] % 2 != 0:
            flag = True
    return flag
    
if check_2(a_ls):
    print(0)
else:
    res = 0
    while True:
        if check_2(a_ls):
            break
        res += 1
        a_ls = [i/2 for i in a_ls]
    print(res)
```
