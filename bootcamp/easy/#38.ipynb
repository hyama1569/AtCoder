```python
import math

ls = []
for i in range(5):
    ls.append(int(input()))
ls_1 = [int(str(ls[i])[-1]) for i in range(len(ls))]

cand_i = 0
cand = ls_1[cand_i]
for i in range(5):
    if ls_1[i] != 0 and ls_1[i] < cand:
        cand_i = i

res = 0
for i in range(5):
    if i == cand_i:
        res += ls[i]
    else:
        res += math.ceil(ls[i]/10)*10
print(res)
```
