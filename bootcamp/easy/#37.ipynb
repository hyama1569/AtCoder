```python
n, k, q = map(int, input().split())
scores = [k-q for i in range(n)]
seikai_num = [0 for i in range(n)]
for i in range(q):
    seikai_id = int(input()) - 1
    seikai_num[seikai_id] += 1
for i in range(n):
    scores[i] += seikai_num[i]

for i in range(n):
    if scores[i] > 0:
        print("Yes")
    else:
        print("No")
```
