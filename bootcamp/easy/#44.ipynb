```python
s = input()

s_01 = ""
s_10 = ""
for i in range(len(s)):
    if i % 2 == 0:
        s_01 += '0'
        s_10 += '1'
    else:
        s_01 += '1'
        s_10 += '0'

res01 = 0
res10 = 0
for i in range(len(s)):
    if s[i] != s_01[i]:
        res01 += 1
    if s[i] != s_10[i]:
        res10 += 1

print(min(res01, res10))
```
