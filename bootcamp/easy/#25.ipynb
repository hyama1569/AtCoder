```python
import itertools

n = int(input())
p_ls = list(map(int, input().split()))
q_ls = list(map(int, input().split()))
perms = list(itertools.permutations(sorted(p_ls), n))
for i in range(len(perms)):
    if perms[i] == tuple(p_ls):
        a = i+1
    if perms[i] == tuple(q_ls):
        b = i+1
print(abs(a - b))
```
