```python
import math
def dist(x1, y1, x2, y2):
    return math.sqrt((x1 - x2)**2 + (y1 - y2)**2)

x1, y1, x2, y2 = map(int, input().split())
diffx = x2 - x1
diffy = y2 - y1
d = dist(x1, y1, x2, y2)
sin = diffy / d
cos = diffx / d

x3 = d * (cos - sin)
y3 = d * (sin + cos)
x4 = d * (- sin)
y4 = d * (cos)

print("{} {} {} {}".format(int(x3+x1), int(y3+y1), int(x4+x1), int(y4+y1)))
```
