```python
n, a, b = map(int, input().split())
s = input()
n_clear = 0
b_rank = 1
for i in range(n):
    if s[i] == 'c':
        print("No")

    elif s[i] == 'a' and n_clear < (a+b):
        print("Yes")
        n_clear += 1

    elif s[i] == 'b' and n_clear < (a+b) and b_rank <= b:
        print("Yes")
        n_clear += 1
        b_rank += 1
        
    else:
        print("No")
```
