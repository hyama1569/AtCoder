```python
n, k = map(int, input().split())
if n < k:
    print(min(n, k - n))
elif n == k:
    print(0)
else:
    print(min(k % (n - k), n % k))
```

```python
#N<=10^18だから愚直に引き算するとダメ
#%を使って予めNに対し、引き算が十分行われた状態にする
n, k = map(int, input().split())

n = n % k
while(True):
    nxt = abs(n - k)
    if nxt < n:
        n = nxt
    else:
        break
print(n)
```
