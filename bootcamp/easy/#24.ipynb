```python
a, b, m = map(int, input().split())
a_ls = list(map(int, input().split()))
b_ls = list(map(int, input().split()))
res = 200001
for i in range(m):
    x, y, c = map(int, input().split())
    res = min(a_ls[x-1] + b_ls[y-1] - c, res)
a_ls = sorted(a_ls, reverse=False)
b_ls = sorted(b_ls, reverse=False)
res = min(res, a_ls[0] + b_ls[0])
print(res)
```
