```python
a, b, c = map(int, input().split())
res = 0
if a%2 != 0 or b%2 != 0 or c%2 != 0:
    print(0)
    
elif a == b and b == c and c == a:
    print(-1)

else:
    while True:
        if a%2 != 0 or b%2 != 0 or c%2 != 0:
            break
        ha = int(a/2)
        hb = int(b/2)
        hc = int(c/2)
        a = hb + hc
        b = hc + ha
        c = ha + hb
        res += 1

    print(res) 
```
