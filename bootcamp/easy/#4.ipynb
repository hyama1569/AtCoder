```python
import math

n = int(input())
st = int(n // 1.09)
en = int(n // 1.07)
flag = False
for i in range(st, en+1):
    if math.floor(i * 1.08) == n:
        flag = True
        print(i)
        break
if flag == False:
    print(":(")
```

```python
# 全探索が安全
import math
n = int(input())
flag = False
for i in range(1, 50001):
    if math.floor(i * 1.08) == n:
        flag = True
        print(i)
        break
if flag == False:
    print(":(")
```
