```python
s = input()
res = 0
for i in range(len(s)):
    tmp = 0
    if s[i] == 'A' or s[i] == 'C' or s[i] == 'G' or s[i] == 'T':
        for j in range(i, len(s)):
            if s[j] == 'A' or s[j] == 'C' or s[j] == 'G' or s[j] == 'T':
                tmp += 1
            else:
                break
    res = max(res, tmp)
print(res)
```
