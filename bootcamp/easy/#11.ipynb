```python
n = int(input())
cands = [2**i for i in range(7)]
for i in range(len(cands)):
    if n < cands[i]:
        print(cands[i-1])
        break
    elif n == cands[i]:
        print(cands[i])
        break
    elif n > cands[i] and n > 64:
        print(64)
        break
    
```
