```python
a, b = map(int, input().split())
s = input()
flag1 = False
flag2 = True
for i in range(len(s)):
    if i == a:
        if s[i] == '-':
            flag1 = True
    else:
        if s[i] == '-':
            flag2 = False
if flag1 == True and flag2 == True:
    print("Yes")
else:
    print("No")    
```
