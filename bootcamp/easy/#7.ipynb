```python
a = [list(map(int, input().split())) for _ in range(3)]
n = int(input())

flags = [list([False, False, False]) for _ in range(3)]

for i in range(n):
    b = int(input())
    for i in range(3):
        for j in range(3):
            if b == a[i][j]:
                flags[i][j] = True

c1 = [flags[0][0], flags[0][1], flags[0][2]]
c2 = [flags[1][0], flags[1][1], flags[1][2]]
c3 = [flags[2][0], flags[2][1], flags[2][2]]

c4 = [flags[0][0], flags[1][0], flags[2][0]]
c5 = [flags[0][1], flags[1][1], flags[2][1]]
c6 = [flags[0][2], flags[1][2], flags[2][2]]

c7 = [flags[0][0], flags[1][1], flags[2][2]]
c8 = [flags[0][2], flags[1][1], flags[2][0]]

if all(c1) or all(c2) or all(c3) or all(c4) or all(c5) or all(c6) or all(c7) or all(c8):
    print("Yes")
else:
    print("No")
```
