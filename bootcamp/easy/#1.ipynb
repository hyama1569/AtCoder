```python
a, b = map(int, input().split())
q = b // a
r = b % a
if r > 0:
    res = q + 1
    print(res)
else:
    print(q)
```

```python
a, b = map(int, input().split())
res = 0
while True:
    num = res * a - (res - 1)
    if num >= b:
        break
    res += 1
print(res)
```
