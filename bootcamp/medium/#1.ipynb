```python
n = int(input())
ls = []
for i in range(n):
    ls.append(int(input()) - 1)
seen = [0 for i in range(n)]

res = 0
now = 0
seen[now] = 1
flag = False
while now != 1:
    next = ls[now]
    if seen[next] == 1:
        flag = True
        break
    seen[next] = 1
    now = next
    res += 1

if flag:
    print(-1)
else:
    print(res)
```
